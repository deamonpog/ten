```python
import pandas as pd
import gzip
import json
import glob
```

```python
dir = 'C:\\STUFF\\RESEARCH\\TwitterLookup\\tlfr\\Cache\\'
```

```python
filestyle = 'querylist_cache_*_*.json.gzip'
```

```python
files = glob.glob(dir + filestyle)
len(files)
```

```python
def ReadFromGzipJSON(in_file_path):
    with gzip.open(in_file_path, 'r') as fin:
        json_bytes = fin.read()
    json_str = json_bytes.decode('utf-8')
    data = json.loads(json_str)
    return data
```
